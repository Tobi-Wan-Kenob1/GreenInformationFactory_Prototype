```python
# Project: GreenInformationFactory
# Notebook: preprocessing.ipynb
# Purpose: Preprocess raw experimental dataset from Zenodo (DOI: 10.5281/zenodo.16256961)

import os
import pandas as pd
import numpy as np
import requests
from zipfile import ZipFile
from io import BytesIO
```

# Download Raw Data

```python
# Define Zenodo record ID and base url

record_id ='16256961'
zenodo_url = f'https://zenodo.org/api/records/{record_id}'

#locad directory
data_dir = '.../data/raw'
os.makedirs(data_dir, exist_ok=True)

#Get the metadata:
print("🔍 Fetching metadata from Zenodo...")
response = requests.get(zenodo_url)
if response.status_code != 200:
    raise Exception(f"❌ Failed to fetch metadata: {response.status_code}")

metadata = response.json()
files = metadata.get("files", [])

# Download the first ZIP file found (or raise if none found)
file_url = None
for f in files:
    if f["key"].endswith(".zip"):
        file_url = f["links"]["self"]
        filename = f["key"]
        break

if not file_url:
    raise Exception("❌ No ZIP file found in Zenodo record.")

# Download the ZIP file
print(f"⬇️ Downloading file: {filename}")
r = requests.get(file_url)
zip_path = os.path.join(data_dir, filename)
with open(zip_path, "wb") as f:
    f.write(r.content)

# Extract contents
print("📦 Extracting ZIP contents...")
with ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(data_dir)

print(f"✅ Files extracted to: {data_dir}")
```

```python
# Define Zenodo record ID and base url
import os
import requests
from zipfile import ZipFile

record_id = '16256961'
#zenodo_url = f'https://zenodo.org/records/{record_id}'
zenodo_url = "https://zenodo.org/records/16256961"
# Load directory - fixed path syntax
data_dir = './data/raw'  # Use relative path instead of '.../data/raw'
os.makedirs(data_dir, exist_ok=True)

# Add proxy configuration if needed
proxies = {
    'http': None,  # Add your proxy here if needed, e.g., 'http://user:pass@proxy:port'
    'https': None  # Add your proxy here if needed, e.g., 'https://user:pass@proxy:port'
}

# Get the metadata with timeout and proxy settings
print("🔍 Fetching metadata from Zenodo...")
try:
    response = requests.get(zenodo_url, proxies=proxies, timeout=30)
    response.raise_for_status()  # Raise exception for 4XX/5XX responses
except requests.exceptions.RequestException as e:
    print(f"❌ Connection error: {e}")
    print("Try checking your internet connection or proxy settings.")
    raise

metadata = response.json()
files = metadata.get("files", [])

# Download the first ZIP file found (or raise if none found)
file_url = None
for f in files:
    if f["key"].endswith(".zip"):
        file_url = f["links"]["self"]
        filename = f["key"]
        break

if not file_url:
    raise Exception("❌ No ZIP file found in Zenodo record.")

# Download the ZIP file with error handling
print(f"⬇️ Downloading file: {filename}")
try:
    r = requests.get(file_url, proxies=proxies, timeout=60)
    r.raise_for_status()
    zip_path = os.path.join(data_dir, filename)
    with open(zip_path, "wb") as f:
        f.write(r.content)
except requests.exceptions.RequestException as e:
    print(f"❌ Download failed: {e}")
    raise

# Extract contents
print("📦 Extracting ZIP contents...")
with ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(data_dir)

print(f"✅ Files extracted to: {data_dir}")
```
